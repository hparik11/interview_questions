```python
def find_number_of_nulls(column, dataframe):

    return sum([1 for each in list(dataframe[column].isnull()) if each is True])
```

```python
## Before imputting
```

```python
# print(f"Column --> {}, nulls: {find_number_of_nulls('', combined_heart_disease_dataframe})

print(find_number_of_nulls('ChestPain', combined_heart_disease_dataframe))
print(find_number_of_nulls('RestECG', combined_heart_disease_dataframe))
print(find_number_of_nulls('Ca', combined_heart_disease_dataframe))
print(find_number_of_nulls('Thal', combined_heart_disease_dataframe))

```
